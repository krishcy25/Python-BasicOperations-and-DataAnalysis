```python
#Topic 2: Conversion of operations/code from SQL to Python

#Code Block 2.1: SQL (where clause with single condition)-->Python Code

#Select all rows where Region='Asia'
## In pandas you can extract the specific column using the column name, in case of numpy- we need to know the exact column location

reg_fl= df['Region']=='Asia'

df_asia_reg=df[reg_fl]

print(df_asia_reg)
```

```python
print(len(df_asia_reg)) # 136 is the output

print(df_asia_reg.shape) # (136,14) is the output
```

```python
#Code Block 2.2: SQL (where clause with multiple conditions)-->Python Code

# Select all rows where Region='Asia' and Sales Channel is Online'
# Using AND (&) operator
# OR Operator can be used like |

reg_fl=df['Region']=='Asia'

sales_ch_fl=df['Sales Channel']== 'Online'

combined= reg_fl & sales_ch_fl

df_Asia_Online=df[combined]

print(df_Asia_Online)
```

```python
#Code Block 2.3: SQL (where clause with multiple conditions using NOT IN)-->Python Code

##Select all rows where Region='Asia' and Sales Channel is not Online'
# Using NOT (!=) operator

reg_fl=df['Region']=='Asia'

sales_ch_fl=df['Sales Channel'] != 'Online'

combined= reg_fl & sales_ch_fl

df_Asia_not_Online=df[combined]

print(df_Asia_not_Online)
```

```python
#Code Block 2.4: SQL (where clause with order by on single variable)-->Python Code

##Select all rows where Region='Asia' order by Unit Cost

reg_fl=df['Region']=='Asia'

df_asia_unit_cost=df[reg_fl].sort_values(by='Unit Cost')

print(df_asia_unit_cost)
```

```python
#Code Block 2.5: SQL (where clause with order by on multiple variables)-->Python Code

##Select all rows where Region='Asia' order by Unit Cost and Units Sold

reg_fl=df['Region']=='Asia'

df_asia_unit_cost_usold=df[reg_fl].sort_values(by=['Unit Cost', 'Units Sold'])

print(df_asia_unit_cost_usold)
## Please note: If you want to have the columns sorted in descending order use like below

## df_asia_unit_cost_usold=df[reg_fl].sort_values(by=['Unit Cost', 'Units Sold'], ascending='False')
```

```python
#4.9 Map funciton
# Map()- this applies the given function to each item of iterable and returns list of results

def calculateSquare(n):
    return n*n

numbers= (1,2,3,4)

result= map(calculateSquare, numbers)

print(result) #Directly using print result won't work. We need to convert the map result to set result and then print

numbersSquare=set(result)

print(numbersSquare)
```
