```python
import pandas as pd
import numpy as np

# Two most common encodings- UTF-8 and Latin-1

trans1=pd.read_csv("Trans1.csv", encoding="Latin-1")
trans2=pd.read_csv("Trans2.csv", encoding="Latin-1")
Info=pd.read_csv("Info.csv", encoding="Latin-1")
```

```python
# Topic 1: Data Combine

#1.1 Combining dataframes using concat function- concat_axis=0
# Without specifying the axis, the default axis=0 (which bascially combines the data vertically like SQL UNION)
Trans=pd.concat([trans1, trans2])

Trans.head(20)
```

```python
#1.2 Using the same above with Ignore Index option
#ignore_index can be used to clear the existing index in the result
#When combining the datasets using above, the indexes in both will remain same. 
#first dataset- index no starts with 0-12 and second starts with 0-2
#On using the ignore_index='True' index numbers continues from 0-15 in the order when data combines vertically. 

Trans=pd.concat([trans1, trans2],ignore_index='True')

Trans.head(20)
```

```python
#1.4 Combining dataframes using Merge function (Inner Join)
# This bascially is similar to inner join where the two customers who exist in Info didn't appear because they dont have transactions
# As we have name variable in both datasets, on combining python creates automatically name_x and name_y in the final dataset
# If we don't submit how, its inner join by default

merge1=pd.merge(left=Info, right=Trans, on='Id')
merge1.head(20)
```

```python
#1.5 Combining dataframes using Merge function (left Join)
# Now you can see 106 and 107 customers who doesn't have transactions
merge_left=pd.merge(left=Info, right=Trans, how='left',on='Id')

merge_left.head()
```

```python
#1.6 Combining dataframes using Merge function (right Join)

merge_right=pd.merge(left=Info, right=Trans, how='right',on='Id')
merge_right.head()
```

```python
#1.7 Combining dataframes using Merge function (outer Join)

merge_outer=pd.merge(left=Info, right=Trans, how='outer',on='Id')

merge_outer.head(20)
```

```python
#1.8 Use of suffixes
# If the variable exists in both datasets during merging, python creates them as var_x and var_y by default
#However, if we want to specify the suffix explicitly- we can use the suffix option like below


merge_suffix=pd.merge(left=Info, right=Trans, on='Id', suffixes=('_d1','_d2'))

merge_suffix.head()
```

```python
## Please Note: We can also join on indexes if needed. However, we won't need that as indexes concept is not useful in the pandas
## Below is the syntax if we want to join the dataset using indexes rather than on any columns
# Only 0-6 rows exits in Info, as we used left join only 0-6 rows from Trans is pulled based on index join

merge_index_suffix=pd.merge(left=Info, right=Trans, how='inner',left_index=True, right_index=True,suffixes=('_d1','_d2'))

merge_index_suffix.head(20)
```

```python
# Topic 2: Transforming Data with Pandas
# Using map(), apply(), applymap(), apply(), melt()


#2.1 Creation of new column based on cases (this is more like CASE statement in SAS)
# Below we will use the Gender column to create a new calculated column in the Info dataset 
# which defines M=1 and F=0

def label(element):
    if element=='M':
        return 1
    else:
        return 0

Info['Gender_Calc']=Info['Gender'].map(label)

# Even the below function works same as map and it gives you the same result 
#Info['Gender_Calc']=Info['Gender'].apply(label)

Info.head()
```

```python
#2.2 Difference between apply() and map()
# Map- works with only pre-definied values. We cannot pass any value in arguments for case statements consideration
# Apply- works with values as well where we can pass values in arguments for case statements consideration
# Below, we use Info data (Income column) and if the income is >50 we will assign High else low

Info_appmap=pd.read_csv("Info.csv", encoding="Latin-1")


def label(element,x):
    if element > x:
        return 'High'
    else:
        return 'Low'

# Here, we use the same case statement definition above with 2 values- 40, 50 (by creating 2 columns) with single case definition
# Using map we cannot do this because map doesn't accept values.

Info_appmap['Income>50']=Info['Income'].apply(label, x=50)

Info_appmap['Income>40']=Info['Income'].apply(label, x=40)


Info_appmap.head()
```

```python
#2.3 Use of applymap()
# Applymap() cannot be applied on a single series. It should be applied to Dataframe only which involves more than 1 column
# This is useful when you want to apply the same case definition calculation to more than 1 column
# For example: lets convert both id and income column in info dataset (>102- High, low)

def label(element):
    if element=='M':
        return 1
    else:
        return 0
        
factors=['State', 'Gender']


#Here we are calculating case statetment definitions for 2 columns. Hence, we cannot give Info_app['One var'] allocation
# as python doesn't know which variable assignment/name is among them
#hence, we need to first define one variable val (which basically stores calculated columns as dataframe)
# later, they can be assigned back to original dataset by pulling columns from intermediate dataframe that was created.
val=Info_appmap[factors].applymap(label)

Info_appmap['Gender_Calc']= val['Gender']
Info_appmap['State_Calc']= val['State']

Info_appmap.head()
## Please note that you can also use apply() function to the dataframe as well which is not covered here
```

```python
#2.4 Using pd.melt(): unpivots a Dataframe from wide format to long format
# Info_appmap has 4 calculated varaibles (Income>50, Income>40, Gender_Calc, and State_Calc)
# Syntax: pd.melt(df,id_vars=[col1,col2],value_vars=[col1,col2])
# id_vars= are the columns that should remain same in the final dataset
#value_vars= are the names of the columns that should be changed to rows in the result
# Melt function is used to reshape our data so that we could use vectorized operation on single column if needed.


id_variables=['Id','Name','State','Country','Gender','Income']
value_variables=['Income>50','Income>40','Gender_Calc','State_Calc']

melt_df=pd.melt(Info_appmap, id_vars=id_variables, value_vars=value_variables)

print(melt_df)
```

```python
#3.3 Calculating the lenth of a string for one column and store it in a different column
 # When calculating the length of any string column, we need to make sure that we take care of NaN values
 # because NaN values will be coming as length as 3 eventhough we knew they are 

def compute_lengths(element):
     if pd.isnull(element):
         pass
     else:
         return len(str(element))     
    
Info_appmap['Country_Length']= Info_appmap['Country'].apply(compute_lengths)


Info_appmap.head()

#You can see that the Country length is calculated which is the last column in the dataframe
```

```python
#3.4 Creating a new calculated column by converting the string stored in one column to upper case letters

Info_appmap['New_Country']= Info_appmap['Country'].str.upper()
Info_appmap.tail()
```

```python
#3.5 Pattern Searching- Using contains
# For example: lets search for pattern in Name in Info_appmap- we are looking for pattern in names that starts with Pa
#Pa_name_pattern column created at the end with True or false based on existing pattern available for Name column

pattern=r"[Pp]a"

Info_appmap['Pa_name_pattern']=Info_appmap['Name'].str.contains(pattern)
Info_appmap.head()
```

```python
# The above column yields us results like True, False (bool_values)
#Although, they are created as True/False, their internal values is still 1 and 0.
# Using True/False in below case definition won't work. We need to use 1 and 0
# Column created with "String values" instead of True/False Boolean values

def pattern_match(element):
    if element==1:
        return "Pattern_Match"
    else:
        return "Pattern_Mismatch"

Info_appmap['Pattern_Match']=Info_appmap['Pa_name_pattern'].apply(pattern_match)
Info_appmap.head()
```

```python
# Topic 4: Working with Missing and Duplicate Data

#4.1 Identifying missing values
# Let's identify missing values in the dataset Info
# The below code prints the column names of the dataset and the number of missing values each column contain

print(Info.isnull().sum())
```

```python
#4.2 Indentifying the duplicate values
# First match in the column will be taken as False(not dup) and later if we have same then it returns True which means dup
print(Info.duplicated(['Country']))
```

```python
#Name is unique in our dataset. Hence, everything should return False as we used name and country combination for dup check
print(Info.duplicated(['Name','Country']))
```

```python
#4.3 Dropping the duplicates
#Info and Info_drop1 will be same after the below statement because there are no duplicates identified to drop

Info_drop1=Info.drop_duplicates(['Name','Country'])
Info_drop1.head()
```

```python
#Let's try with only country where we identified the duplicates
# only 2 rows should be returned because 5 rows are identified as duplicates
Info_drop2=Info.drop_duplicates(['Country'])
Info_drop2.head()
```
